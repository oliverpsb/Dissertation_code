```python
import pandas as pd
```

```python
factors_df = pd.read_csv('all_factors.csv')
monthly_macro_df = pd.read_csv('monthly_macro.csv')
```

```python
factors_df['Date'] = pd.to_datetime(factors_df['Date'], format='%d/%m/%Y')
monthly_macro_df['Date'] = pd.to_datetime(monthly_macro_df['Date'], format="%Y-%m-%d")
```

```python
# Inner join to keep only matching dates
merged_df = pd.merge(factors_df, monthly_macro_df, on='Date', how='inner')
```

```python
merged_df.to_csv('merged_df.csv', index=False)
```
